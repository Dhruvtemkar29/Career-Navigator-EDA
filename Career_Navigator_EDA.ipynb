```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler

df = pd.read_csv('education_career_success.csv')
print(df.head(3))
```

```python
print("Rows and Columns :", df.shape)

print("\nColumns:")
print(df.columns)

print("\nData Types:")
print(df.dtypes)
```

```python
print(df.isnull().sum())
```

```python
print(df.duplicated().sum())
```

```python
print(df.describe())
```

```python
numerical_cols = [
    'Age',
    'High_School_GPA',
    'SAT_Score',
    'University_GPA',
    'Internships_Completed',
    'Projects_Completed',
    'Certifications',
    'Soft_Skills_Score',
    'Networking_Score',
    'Job_Offers',
    'Starting_Salary',
    'Career_Satisfaction',
    'Years_to_Promotion',
    'Work_Life_Balance'
]
```


---



**Before and After Normalisation**


---



```python
minmax = MinMaxScaler()

df_minmax = df.copy()

df_minmax[numerical_cols] = minmax.fit_transform(
    df[numerical_cols]
)

df_minmax.head()
```

```python
standard = StandardScaler()

df_zscore = df.copy()

df_zscore[numerical_cols] = standard.fit_transform(
    df[numerical_cols]
)

df_zscore.head()
```

```python
comparison = pd.DataFrame({
    'Original': df['Starting_Salary'].head(),
    'Min-Max': df_minmax['Starting_Salary'].head(),
    'Z-Score': df_zscore['Starting_Salary'].head()
})

comparison
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.histplot(df['Starting_Salary'], kde=True)
plt.title('Original Salary')

plt.subplot(1,2,2)
sns.histplot(df_zscore['Starting_Salary'], kde=True)
plt.title('Z-Score Normalized Salary')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.boxplot(y=df['Starting_Salary'])
plt.title('Original Salary')

plt.subplot(1,2,2)
sns.boxplot(y=df_minmax['Starting_Salary'])
plt.title('Min-Max Salary')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.boxplot(y=df['Starting_Salary'])
plt.title('Original Salary')

plt.subplot(1,2,2)
sns.boxplot(y=df_zscore['Starting_Salary'])
plt.title('Z-Score Salary')

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(1,2,figsize=(15,5))

sns.barplot(
    x='Field_of_Study',
    y='Starting_Salary',
    data=df,
    ax=ax[0]
)

ax[0].set_title('Original Salary')

sns.barplot(
    x='Field_of_Study',
    y='Starting_Salary',
    data=df_minmax,
    ax=ax[1]
)

ax[1].set_title('Min-Max Salary')

for a in ax:
    a.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(1,2,figsize=(14,5))

sns.lineplot(
    x='Networking_Score',
    y='Starting_Salary',
    data=df,
    ax=ax[0]
)

ax[0].set_title('Original Data')

sns.lineplot(
    x='Networking_Score',
    y='Starting_Salary',
    data=df_minmax,
    ax=ax[1]
)

ax[1].set_title('Min-Max Data')

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(1,2,figsize=(16,6))

sns.heatmap(
    df[numerical_cols].corr(),
    annot=True,
    cmap='coolwarm',
    ax=ax[0]
)

ax[0].set_title('Original Data')

sns.heatmap(
    df_minmax[numerical_cols].corr(),
    annot=True,
    cmap='coolwarm',
    ax=ax[1]
)

ax[1].set_title('Min-Max Data')

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(1,3,figsize=(18,5))

sns.histplot(df['Starting_Salary'], kde=True, ax=ax[0])
ax[0].set_title('Original')

sns.histplot(df_minmax['Starting_Salary'], kde=True, ax=ax[1])
ax[1].set_title('Min-Max')

sns.histplot(df_zscore['Starting_Salary'], kde=True, ax=ax[2])
ax[2].set_title('Z-Score')

plt.tight_layout()
plt.show()
```
